```python
#W4A10
from math import *
n = int(input())
a = [0] * (n + 1)
def collazt (x):
  step = 1
  while x != 1:
    if x % 2 == 0:
      x //= 2
    else:
      x = 3 * x + 1
    step +=1
  return step
for i in range (1,n+1):
  a[i] = collazt(i)
max_length = 0
index_x = 0
for k in range (1, n+1):
  if a[k] >= max_length:
    max_length = a[k]
    index_x = k
print(index_x, max_length)
```
