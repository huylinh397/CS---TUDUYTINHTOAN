```python
def hoan_vi_tiep_theo(m,lst):
  n = m-1
  i = n-1
  while True:
    if lst[i] >= lst[i+1]:
      i-=1
    else:
      break
  if n < 0:
    lst = lst.reseve()
  for j in range(n,-1, -1):
    if lst[j]>lst[i]:
      break
  lst[i], lst[j] = lst[j], lst[i]
  lst[i+1:] = reversed(lst[i+1:])
  return lst
hoan_vi_tiep_theo(6,[4,5,6,3,2,1])

```

```python
def is_path_crossing(moves):
  i, j = 0, 0
  d=[]
  for x in moves:
    if x =='R':
      i+=1
    if x =='L':
      i-=1
    if x =='U':
      j+=1
    if x =='D':
      j-=1
    if [i,j] in d:
      return True
    else:
      d.append([i,j])
  return False
print(is_path_crossing("RURDL"))

```
