```python
#W7A1
def binary_search(a, l, r, x):
  res = -1
  while l <= r:
    m = (l + r) // 2
    if a[m] == x:
      res = m
    if a[m] < x:
      l = m + 1
    else:
      r = m - 1
  return res
a = [1,2,3,3,4,5,6,7,8]
l = a[0]
r = a[len(a)-1]
x = 9
print(binary_search(a, l, r, x))

```

```python
#W7A3
def bubble_sort(arr):
  n = len(arr)
  for i in range (n):
    swapped = False
    for j in range(0, n-i-1):
      if arr[j] > arr[j+1]:
        arr[j], arr[j+1] = arr[j+1], arr[j]
        swapped = True
      if not swapped:
        break

```
